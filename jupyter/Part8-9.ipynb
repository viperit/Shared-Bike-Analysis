```python
def get_to_end_ratio(to_end_count_path,user_count_path='./data/train_hard.csv'):
    # 一个是终点到达次数计算的csv的路径，一个是计算用户使用了几次的path
    # 在终点飘逸的时候，user_count_path用户使用的次数不能是飘逸的
    
    user_count = {}# 用户出发次数==用户行程次数
    user_hour_count = {}# 用户每小时出发次数
    i = 0
    csv_reader = csv.reader(open(user_count_path, encoding='utf-8'))
    for row in csv_reader:
        i+=1
        if i == 1:
            continue
        user = row[1]
        hour = row[7]
        
        if user not in user_count:
            user_count[user] = 1
        else:
            user_count[user] += 1
        
        if user not in user_hour_count:
            user_hour_count[user] = {}
            user_hour_count[user][hour] = 1
        else:
            if hour not in user_hour_count[user]:
                user_hour_count[user][hour] = 1
            else:
                user_hour_count[user][hour] += 1
    
    user_cur_hour_loc_as_end = {}# 该用户该小时去该终点次数
    
    i = 0
    csv_reader = csv.reader(open(to_end_count_path, encoding='utf-8'))
    for row in csv_reader:
        i+=1
        if i == 1:
            continue
        
        user = row[1]
        end = row[6]
        hour = row[7]
        
        if user not in user_cur_hour_loc_as_end:
            user_cur_hour_loc_as_end[user] = {}
            user_cur_hour_loc_as_end[user][hour] = {}
            user_cur_hour_loc_as_end[user][hour][end] = 1
        else:
            if hour not in user_cur_hour_loc_as_end[user]:
                user_cur_hour_loc_as_end[user][hour] = {}
                user_cur_hour_loc_as_end[user][hour][end] = 1
            else:
                if end not in user_cur_hour_loc_as_end[user][hour]:
                    user_cur_hour_loc_as_end[user][hour][end] = 1
                else:
                    user_cur_hour_loc_as_end[user][hour][end] += 1
                    
    user_cur_hour_loc_as_end_ratio = {}# 该用户该小时去该终点／用户总出发次数

    user_cur_hour_loc_as_end_cur_hour_ratio = {}# 该用户该小时去该终点／用户该小时出发次数
    
    for user in user_cur_hour_loc_as_end:
        user_cur_hour_loc_as_end_ratio[user] = {}
        user_cur_hour_loc_as_end_cur_hour_ratio[user] = {}
        
        for hour in user_cur_hour_loc_as_end[user]:
            user_cur_hour_loc_as_end_ratio[user][hour] = {}
            user_cur_hour_loc_as_end_cur_hour_ratio[user][hour] = {}
            
            for end in user_cur_hour_loc_as_end[user][hour]:
                user_cur_hour_loc_as_end_ratio[user][hour][end] = \
                                                user_cur_hour_loc_as_end[user][hour][end]/user_count[user]
                user_cur_hour_loc_as_end_cur_hour_ratio[user][hour][end] = \
                                                user_cur_hour_loc_as_end[user][hour][end]/user_hour_count[user][hour]
            
            
    return user_cur_hour_loc_as_end_ratio,user_cur_hour_loc_as_end_cur_hour_ratio
```

```python
user_cur_hour_loc_as_end_ratio,user_cur_hour_loc_as_end_cur_hour_ratio = get_to_end_ratio('./data/train_hard.csv')

user_cur_hour_loc_as_end_geofly_ratio,user_cur_hour_loc_as_end_geofly_cur_hour_ratio\
                                    = get_to_end_ratio('./data/train_end_geofly.csv')
```

```python
def get_hour_to_end_ratio_s(user_count_path='./data/train_hard.csv'):
    # 一个是当前小时终点到达次数计算的csv的路径，一个是计算用户使用了几次的path
    # 在终点飘逸的时候，user_count_path用户使用的次数不能是飘逸的
    neighbour = {}
    i = 0
    csv_reader = csv.reader(open('./data/temp/neighbour.csv', encoding='utf-8'))
    for row in csv_reader:
        i+=1
        if i == 1:
            continue
        neighbour[row[0]] = eval(row[1])
    
    user_count = {}# 用户出发次数==用户行程次数
    user_hour_count = {}# 用户每小时出发次数
    i = 0
    csv_reader = csv.reader(open(user_count_path, encoding='utf-8'))
    for row in csv_reader:
        i+=1
        if i == 1:
            continue
        user = row[1]
        hour = row[7]
        
        if user not in user_count:
            user_count[user] = 1
        else:
            user_count[user] += 1
        
        if user not in user_hour_count:
            user_hour_count[user] = {}
            user_hour_count[user][hour] = 1
        else:
            if hour not in user_hour_count[user]:
                user_hour_count[user][hour] = 1
            else:
                user_hour_count[user][hour] += 1
        
    
    # 记录当前小时该用户从该起点到该终点的次数
    user_cur_hour_start_to_end = {}
    
    i = 0
    csv_reader = csv.reader(open(user_count_path, encoding='utf-8'))
    for row in csv_reader:
        i+=1
        if i == 1:
            continue
        
        user = row[1]
        starts = neighbour[row[5]]
        ends = neighbour[row[6]]
        hour = row[7]
        
        for start in starts:
            for end in ends:
                if user not in user_cur_hour_start_to_end:
                    user_cur_hour_start_to_end[user] = {}
                    user_cur_hour_start_to_end[user][hour] = {}
                    user_cur_hour_start_to_end[user][hour][start] = {}
                    user_cur_hour_start_to_end[user][hour][start][end] = 1
                else:
                    if hour not in user_cur_hour_start_to_end[user]:
                        user_cur_hour_start_to_end[user][hour] = {}
                        user_cur_hour_start_to_end[user][hour][start] = {}
                        user_cur_hour_start_to_end[user][hour][start][end] = 1
                    else:
                        if start not in user_cur_hour_start_to_end[user][hour]:
                            user_cur_hour_start_to_end[user][hour][start] = {}
                            user_cur_hour_start_to_end[user][hour][start][end] = 1
                        else:
                            if end not in user_cur_hour_start_to_end[user][hour][start]:
                                user_cur_hour_start_to_end[user][hour][start][end] = 1
                            else:
                                user_cur_hour_start_to_end[user][hour][start][end] += 1
    
                                
    
                    
    user_cur_hour_start_to_end_ratio = {}# 该用户该小时从此起点到终点／用户总出发次数

    user_cur_hour_start_to_end_cur_hour_ratio = {}# 该用户该小时从此起点到终点点／用户该小时出发次数
    
    for user in user_cur_hour_start_to_end:
        user_cur_hour_start_to_end_ratio[user] = {}
        user_cur_hour_start_to_end_cur_hour_ratio[user] = {}
        
        for hour in user_cur_hour_start_to_end[user]:
            user_cur_hour_start_to_end_ratio[user][hour] = {}
            user_cur_hour_start_to_end_cur_hour_ratio[user][hour] = {}
            
            for start in user_cur_hour_start_to_end[user][hour]:
                user_cur_hour_start_to_end_ratio[user][hour][start] = {}
                user_cur_hour_start_to_end_cur_hour_ratio[user][hour][start] = {}
                
                for end in user_cur_hour_start_to_end[user][hour][start]:
                    user_cur_hour_start_to_end_ratio[user][hour][start][end] = \
                                user_cur_hour_start_to_end[user][hour][start][end]/user_count[user]
                        
                    user_cur_hour_start_to_end_cur_hour_ratio[user][hour][start][end] = \
                                user_cur_hour_start_to_end[user][hour][start][end]/user_hour_count[user][hour]
             
            
    return user_cur_hour_start_to_end_ratio,user_cur_hour_start_to_end_cur_hour_ratio
```

```python
def write_large_pkl(file_path,data):
    ## write
    n_bytes = 2**31
    max_bytes = 2**31 - 1
    
    bytes_out = pickle.dumps(data)
    
    with open(file_path, 'wb') as f_out:
        for idx in range(0, n_bytes, max_bytes):
            f_out.write(bytes_out[idx:idx+max_bytes])

def read_large_pkl(file_path):
    n_bytes = 2**31
    max_bytes = 2**31 - 1
    
    bytes_in = bytearray(0)
    input_size = os.path.getsize(file_path)
    
    with open(file_path, 'rb') as f_in:
        for _ in range(0, input_size, max_bytes):
            bytes_in += f_in.read(max_bytes)
    return pickle.loads(bytes_in)
```
